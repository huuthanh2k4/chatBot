```python
import sys
import os
import tempfile
from datetime import datetime, date
import pickle
import numpy as np

import speech_recognition as sr
from gtts import gTTS

from PyQt5.QtWidgets import (
    QApplication, QWidget, QVBoxLayout, QTextEdit, QLineEdit,
    QPushButton, QHBoxLayout, QLabel, QFrame
)
from PyQt5.QtCore import QTimer, QUrl, QEventLoop
from PyQt5.QtMultimedia import QMediaPlayer, QMediaContent


class Waguri_kaoruko_chat(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Waguri kaoruko-Chanüíñüíñ")
        self.setGeometry(200, 200, 450, 550)
        self.setStyleSheet("background-color: #FFC0CB;")
        
        layout = QVBoxLayout()
        
        self.chat_display = QTextEdit()
        self.chat_display.setReadOnly(True)
        self.chat_display.setStyleSheet(
            "font-size: 18px; background-color: white; border-radius: 15px; padding: 15px; color: black;"
        )
        layout.addWidget(self.chat_display)
        
        self.typing_label = QLabel("")
        self.typing_label.setStyleSheet("font-size: 14px; color: red; font-style: italic;")
        layout.addWidget(self.typing_label)
        
        input_frame = QFrame()
        input_frame.setStyleSheet("background-color: white; border-radius: 15px; padding: 5px;")
        input_layout = QHBoxLayout()
        
        self.input_field = QLineEdit()
        self.input_field.setStyleSheet(
            "font-size: 18px; padding: 10px; border: none; background-color: white;"
        )
        input_layout.addWidget(self.input_field)
        
        self.send_button = QPushButton("üíì G·ª≠i")
        self.send_button.setStyleSheet(
            "font-size: 18px; background-color: #FF69B4; color: white; border-radius: 10px; padding: 10px;"
        )
        self.send_button.clicked.connect(self.send_message)
        input_layout.addWidget(self.send_button)
        
        input_frame.setLayout(input_layout)
        layout.addWidget(input_frame)
        
        self.setLayout(layout)
        
        # Dictionary ch·ª©a c√°c t·ª´ kh√≥a v√† ph·∫£n h·ªìi t∆∞∆°ng ·ª©ng
        self.keyword_responses = {
            "bu·ªìn ng·ªß": "C·∫≠u m·ªát l·∫Øm r·ªìi ph·∫£i honggg~? V·∫≠y nh·∫Øm m·∫Øt l·∫°i n√®, ƒë·ªÉ t·ªõ ru c·∫≠u ng·ªß nha~ üò¥üíï",
            "m·ªát": "N·∫øu c·∫≠u m·ªát th√¨ c·∫≠u h√£y ngh·ªâ ng∆°i ƒëi nh√© üòò",
            "ch√°n": "Sao c·∫≠u c·∫£m th·∫•y ch√°n v·∫≠y, t·ªõ lu√¥n s·∫µn s√†ng l·∫Øng nghe c·∫≠u üíñ ‚ù§Ô∏è",
            "y√™u": "T·ªõ iuuu c·∫≠u nh√¨u l·∫Øm √°aa√°aa üíï",
            "vui": "Th·∫≠t tuy·ªát khi l√†m c·∫≠u c·∫£m th·∫•y vuiii üíñüíñ",
            "mu·ªôn phi·ªÅn": "N·∫øu c·∫≠u c·∫£m th·∫•y c√¥ ƒë∆°n, nh·ªõ r·∫±ng t·ªõ lu√¥n ·ªü b√™n c·∫≠u, c√πng nhau ch√∫ng ta s·∫Ω v∆∞·ª£t qua üíï",
            "t√™n": "T·ªõ l√† Waguri kaorukoüíñüíñ, c·∫≠u qu√™n t√™n t·ªõ √† ü•∫",
            "kh√¥ng bi·∫øt n·ªØa": "Kh√¥ng sao c·∫£, c·∫≠u v·∫´n c√≥ t·ªõ ·ªü ƒë√¢y m√†üíñüíñ üòò",
            "kh√¥ng c·∫≠u" : "H·ª´m üò§üí¢",
            "xin l·ªói": "Kh√¥ng sao m√†, t·ªõ kh√¥ng gi·∫≠n ƒë√¢u n√®~ l·∫°i ƒë√¢y ƒë·ªÉ t·ªõ xoa ƒë·∫ßu n√†o! üíïü§ó",
            "d·ªÖ th∆∞∆°ng": "Ehhh~! C·∫≠u n√≥i g√¨ z·∫≠y? ƒê·ª´ng l√†m t·ªõ ng·∫°i ch·ª©! (///À¨///)üíì",
            "ƒë√≥i": "·ªêi z·ªùi ∆°i~~! C·∫≠u ƒë√≥i r·ªìi √†? ƒê·ªÉ t·ªõ n·∫•u g√¨ ƒë√≥ cho c·∫≠u nhaa~ üçôüçúüíñ",
            "mu·ªôn" : "Hoi,mu·ªôn r·ªìi, c·∫≠u n√™n ƒëi ng·ªß s·ªõm ƒëi nha, ng·ªß s·ªõm gi√∫p c∆° th·ªÉ kh·ªèe m·∫°nh l·∫Øm ƒë√≥ üíñüíñ",
            "nh·ªõ": "Aww~ c·∫≠u nh·ªõ t·ªõ h·∫£? T·ªõ c≈©ng nh·ªõ c·∫≠u l·∫Øm n√®~~ l·∫°i √¥m c√°i n√†ooo! ü§óüíï",
            "hello": "Ch√†o c·∫≠u \n H√¥m nay c·∫≠u th·∫ø n√†o v·∫≠y? üòä",
            "ch√†o c·∫≠u": "Ch√†o c·∫≠u \n H√¥m nay c·∫≠u th·∫ø n√†o v·∫≠y? üòä",
            "n√≥i chuy·ªán": "T·ªõ ƒë√¢y, c·∫≠u mu·ªën n√≥i chuy·ªán g√¨ v·ªõi t·ªõ? üòä",
            "hmm": "C·∫≠u ƒëang suy nghƒ© v·ªÅ ƒëi·ªÅu g√¨ hay ? T·ªõ ·ªü ƒë√¢y ƒë·ªÉ l·∫Øng nghe c·∫≠u nha~ üòä",
            "·ªïn" : "N·∫øu c·∫≠u ·ªïn th√¨ t·ªõ c≈©ng vui l·∫Øm r·ªìi üíñüíñ",
        }
        
    def send_message(self):
        user_text = self.input_field.text().strip()
        if user_text:
            # Hi·ªÉn th·ªã tin nh·∫Øn c·ªßa ng∆∞·ªùi d√πng
            self.chat_display.append(f"<b style='color:blue;'>B·∫°n:</b> {user_text}")
            # L∆∞u tin nh·∫Øn (chuy·ªÉn th√†nh ch·ªØ th∆∞·ªùng ƒë·ªÉ so s√°nh v·ªõi t·ª´ kh√≥a)
            self.last_user_text = user_text.lower()
            self.input_field.clear()
            self.typing_label.setText("üíñWaguri kaoruko ƒëang g√µ...")
            QTimer.singleShot(2000, self.respond_to_message)
    
    def respond_to_message(self):
        response = self.get_response(self.last_user_text)
        self.typing_label.setText("")
        self.chat_display.append(f"<b style='color:red;'>Waguri kaorukoüíñüíñ:</b> {response}")
    
    def listen_and_convert(self):
        recognizer = sr.Recognizer()
        with sr.Microphone() as source:
            while True:
                # Hi·ªÉn th·ªã th√¥ng b√°o ngay l·∫≠p t·ª©c
                self.chat_display.append("<i>Waguri kaoruko: ƒêang nghe...</i>")
                QApplication.processEvents()  # C·∫≠p nh·∫≠t giao di·ªán ngay l·∫≠p t·ª©c

                recognizer.adjust_for_ambient_noise(source, duration=1)
                try:
                    audio = recognizer.listen(source)
                    recognized_text = recognizer.recognize_google(audio, language="vi-VN")
                    self.chat_display.append(f"<b style='color:blue;'>B·∫°n (mic):</b> {recognized_text}")
                except sr.UnknownValueError:
                    self.chat_display.append("<i>Waguri kaoruko: Xin l·ªói, kh√¥ng nh·∫≠n d·∫°ng ƒë∆∞·ª£c gi·ªçng n√≥i.</i>")
                    continue
                except sr.RequestError:
                    self.chat_display.append("<i>L·ªói k·∫øt n·ªëi m·∫°ng khi nh·∫≠n d·∫°ng gi·ªçng n√≥i.</i>")
                    continue

                # N·∫øu ng∆∞·ªùi d√πng n√≥i "t·∫°m bi·ªát" ho·∫∑c "bye", tho√°t kh·ªèi ch·∫ø ƒë·ªô mic
                if "t·∫°m bi·ªát" in recognized_text.lower() or "bye" in recognized_text.lower():
                    self.chat_display.append("<i>K·∫øt th√∫c cu·ªôc n√≥i chuy·ªán v·ªõi Waguri kaoruko.</i>")
                    break

                # L·∫•y ph·∫£n h·ªìi d·ª±a tr√™n vƒÉn b·∫£n nh·∫≠n d·∫°ng ƒë∆∞·ª£c
                response_text = self.get_response(recognized_text)
                if response_text:
                    self.chat_display.append(f"<b style='color:red;'>Waguri kaorukoüíñüíñ:</b> {response_text}")

                    played = False
                    # Ki·ªÉm tra xem trong vƒÉn b·∫£n c√≥ ch·ª©a t·ª´ kh√≥a n√†o c·ªßa file √¢m thanh kh√¥ng
                    for keyword in self.keyword_responses:
                        if keyword in recognized_text.lower():
                            file_path = os.path.join("voice", f"{keyword}.wav")
                            if os.path.exists(file_path):
                                self.player = QMediaPlayer()
                                media = QMediaContent(QUrl.fromLocalFile(file_path))
                                self.player.setMedia(media)
                                self.player.play()
                                
                                # Ch·ªù ƒë·∫øn khi ph√°t xong
                                loop = QEventLoop()
                                self.player.stateChanged.connect(
                                    lambda state: loop.quit() if state == QMediaPlayer.StoppedState else None
                                )
                                loop.exec_()
                                
                                played = True
                                break

                    # N·∫øu kh√¥ng t√¨m th·∫•y file √¢m thanh ph√π h·ª£p
                    if not played:
                        # N·∫øu ph·∫£n h·ªìi m·∫∑c ƒë·ªãnh, ph√°t file "ko nghe r√µ.wav"
                        if response_text == "C·∫≠u n√≥i l·∫°i ƒëi, t·ªõ kh√¥ng hi·ªÉu √Ω c·∫≠u l·∫Øm üòÖ":
                            file_path = os.path.join("voice", "ko nghe r√µ.wav")
                            if os.path.exists(file_path):
                                self.player = QMediaPlayer()
                                media = QMediaContent(QUrl.fromLocalFile(file_path))
                                self.player.setMedia(media)
                                self.player.play()
                                
                                # Ch·ªù ƒë·∫øn khi ph√°t xong
                                loop = QEventLoop()
                                self.player.stateChanged.connect(
                                    lambda state: loop.quit() if state == QMediaPlayer.StoppedState else None
                                )
                                loop.exec_()
                            else:
                                # N·∫øu file kh√¥ng t·ªìn t·∫°i th√¨ d√πng gTTS l√†m fallback
                                tts = gTTS(text=response_text, lang='vi')
                                with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
                                    temp_audio = fp.name
                                tts.save(temp_audio)
                                self.player = QMediaPlayer()
                                media = QMediaContent(QUrl.fromLocalFile(temp_audio))
                                self.player.setMedia(media)
                                self.player.play()
                                
                                loop = QEventLoop()
                                self.player.stateChanged.connect(
                                    lambda state: loop.quit() if state == QMediaPlayer.StoppedState else None
                                )
                                loop.exec_()
                        else:
                            # C√°c ph·∫£n h·ªìi kh√°c s·ª≠ d·ª•ng gTTS nh∆∞ b√¨nh th∆∞·ªùng
                            tts = gTTS(text=response_text, lang='vi')
                            with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
                                temp_audio = fp.name
                            tts.save(temp_audio)
                            self.player = QMediaPlayer()
                            media = QMediaContent(QUrl.fromLocalFile(temp_audio))
                            self.player.setMedia(media)
                            self.player.play()
                            
                            # Ch·ªù ƒë·∫øn khi ph√°t xong
                            loop = QEventLoop()
                            self.player.stateChanged.connect(
                                lambda state: loop.quit() if state == QMediaPlayer.StoppedState else None
                            )
                            loop.exec_()





    
    def get_response(self, text):
        # N·∫øu ng∆∞·ªùi d√πng nh·∫≠p t·ª´ "mic" th√¨ chuy·ªÉn sang ch·∫ø ƒë·ªô nghe v√† chuy·ªÉn gi·ªçng n√≥i th√†nh vƒÉn b·∫£n
        if "mic" in text.lower():
            self.listen_and_convert()
            return ""
        
        # C√°c x·ª≠ l√Ω kh√°c (v√≠ d·ª•: h·ªèi gi·ªù, t·ª´ kh√≥a ph·∫£n h·ªìi, ...)
        current_time = datetime.now().strftime("%H:%M:%S")
        if "m·∫•y gi·ªù" in text or "b√¢y gi·ªù l√† m·∫•y gi·ªù" in text:
            current_hour = int(datetime.now().hour)
            if 22 <= current_hour or current_hour < 3:
                return f"B√¢y gi·ªù l√† {current_time} \n\n\n C≈©ng mu·ªôn r·ªìi ƒë√≥ c·∫≠u h√£y ƒëi ng·ªß s·ªõm ƒëi üòä"
            elif 5 <= current_hour < 12:
                return f"B√¢y gi·ªù l√† {current_time} \n\n\n Ch√∫c c·∫≠u m·ªôt ng√†y m·ªõi t·ªët l√†nh üíñ"
            elif 12 <= current_hour < 17:
                return f"B√¢y gi·ªù l√† {current_time} \n\n\n Ch√∫c c·∫≠u m·ªôt ng√†y l√†m vi·ªác hi·ªáu qu·∫£ üí™"
            elif 17 <= current_hour < 22:
                return f"B√¢y gi·ªù l√† {current_time} \n\n\n Ch√∫c c·∫≠u c√≥ m·ªôt bu·ªïi t·ªëi vui v·∫ª üíñ"
            else:
                return f"B√¢y gi·ªù l√† {current_time}"
        
        if "b·ªánh tim m·∫°ch" in text:
            from PyQt5.QtWidgets import QInputDialog
            # Nh·∫≠n d·ªØ li·ªáu t·ª´ ng∆∞·ªùi d√πng qua h·ªôp tho·∫°i
            age, ok = QInputDialog.getInt(self, "Nh·∫≠p tu·ªïi", "Nh·∫≠p tu·ªïi:")
            if not ok:
                return "H·ªßy chu·∫©n ƒëo√°n b·ªánh."
            gender, ok = QInputDialog.getInt(self, "Nh·∫≠p gi·ªõi t√≠nh", "Nh·∫≠p gi·ªõi t√≠nh (0: Nam, 1: N·ªØ):")
            if not ok:
                return "H·ªßy chu·∫©n ƒëo√°n b·ªánh."
            chest_pain, ok = QInputDialog.getInt(self, "Tri·ªáu ch·ª©ng ƒëau ng·ª±c", 
                                                "Nh·∫≠p tri·ªáu ch·ª©ng ƒëau ng·ª±c (1: Typical angina, 0: Asymptomatic, 3: Non-anginal pain, 2: Atypical angina):")
            if not ok:
                return "H·ªßy chu·∫©n ƒëo√°n b·ªánh."
            blood_pressure, ok = QInputDialog.getInt(self, "Nh·∫≠p huy·∫øt √°p", "Nh·∫≠p huy·∫øt √°p:")
            if not ok:
                return "H·ªßy chu·∫©n ƒëo√°n b·ªánh."
            cholesterol, ok = QInputDialog.getInt(self, "Nh·∫≠p cholesterol", "Nh·∫≠p cholesterol:")
            if not ok:
                return "H·ªßy chu·∫©n ƒëo√°n b·ªánh."
            heartbeat, ok = QInputDialog.getInt(self, "Nh·∫≠p nh·ªãp tim", "Nh·∫≠p nh·ªãp tim:")
            if not ok:
                return "H·ªßy chu·∫©n ƒëo√°n b·ªánh."
            thalassemia, ok = QInputDialog.getInt(self, "Nh·∫≠p Thalassemia", "Nh·∫≠p Thalassemia:")
            if not ok:
                return "H·ªßy chu·∫©n ƒëo√°n b·ªánh."

            # T·∫°o m·∫£ng ƒë·∫∑c tr∆∞ng t·ª´ d·ªØ li·ªáu ng∆∞·ªùi d√πng nh·∫≠p v√†o
            ho_so = np.array([[age, gender, chest_pain, blood_pressure, cholesterol, heartbeat, thalassemia]])
            
            # S·ª≠ d·ª•ng model ƒë√£ load t·ª´ cell tr∆∞·ªõc (Bac_si_KNN) ƒë·ªÉ d·ª± ƒëo√°n
            try:
                ket_qua = Bac_si_KNN.predict(ho_so)
            except Exception as e:
                result = f"C√≥ l·ªói khi d·ª± ƒëo√°n: {e}"
                self.chat_display.append(f"<b style='color:red;'>Waguri kaorukoüíñüíñ:</b> {result}")
                return result
            
            # X√°c ƒë·ªãnh k·∫øt qu·∫£ d·ª±a tr√™n d·ª± ƒëo√°n c·ªßa model
            if ket_qua[0] == 0:
                result = 'Ch√∫c m·ª´ng b·∫°n kh√¥ng b·ªã b·ªánh v·ªÅ tim m·∫°ch'
            else:
                result = 'B·∫°n c√≥ nguy c∆° b·ªã b·ªánh v·ªÅ tim m·∫°ch, h√£y tham kh·∫£o √Ω ki·∫øn b√°c sƒ©!!!!!!'
            
            # Hi·ªÉn th·ªã k·∫øt qu·∫£ l√™n giao di·ªán chat
            self.chat_display.append(f"<b style='color:red;'>Waguri kaorukoüíñüíñ:</b> {result}")
            return result

    

        for keyword in self.keyword_responses:
            if keyword in text.lower():
                return self.keyword_responses[keyword]
        
        return "C·∫≠u n√≥i l·∫°i ƒëi, t·ªõ kh√¥ng hi·ªÉu √Ω c·∫≠u l·∫Øm üòÖ"
        
if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = Waguri_kaoruko_chat()
    window.show()
    sys.exit(app.exec_())
```
