```python
import sys
import os
import tempfile
from datetime import datetime, date
import pickle
import numpy as np

import speech_recognition as sr
from gtts import gTTS

from PyQt5.QtWidgets import (
    QApplication, QWidget, QVBoxLayout, QTextEdit, QLineEdit,
    QPushButton, QHBoxLayout, QLabel, QFrame
)
from PyQt5.QtCore import QTimer, QUrl, QEventLoop
from PyQt5.QtMultimedia import QMediaPlayer, QMediaContent


class Waguri_kaoruko_chat(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Waguri kaoruko-Chan💖💖")
        self.setGeometry(200, 200, 450, 550)
        self.setStyleSheet("background-color: #FFC0CB;")
        
        layout = QVBoxLayout()
        
        self.chat_display = QTextEdit()
        self.chat_display.setReadOnly(True)
        self.chat_display.setStyleSheet(
            "font-size: 18px; background-color: white; border-radius: 15px; padding: 15px; color: black;"
        )
        layout.addWidget(self.chat_display)
        
        self.typing_label = QLabel("")
        self.typing_label.setStyleSheet("font-size: 14px; color: red; font-style: italic;")
        layout.addWidget(self.typing_label)
        
        input_frame = QFrame()
        input_frame.setStyleSheet("background-color: white; border-radius: 15px; padding: 5px;")
        input_layout = QHBoxLayout()
        
        self.input_field = QLineEdit()
        self.input_field.setStyleSheet(
            "font-size: 18px; padding: 10px; border: none; background-color: white;"
        )
        input_layout.addWidget(self.input_field)
        
        self.send_button = QPushButton("💓 Gửi")
        self.send_button.setStyleSheet(
            "font-size: 18px; background-color: #FF69B4; color: white; border-radius: 10px; padding: 10px;"
        )
        self.send_button.clicked.connect(self.send_message)
        input_layout.addWidget(self.send_button)
        
        input_frame.setLayout(input_layout)
        layout.addWidget(input_frame)
        
        self.setLayout(layout)
        
        # Dictionary chứa các từ khóa và phản hồi tương ứng
        self.keyword_responses = {
            "buồn ngủ": "Cậu mệt lắm rồi phải honggg~? Vậy nhắm mắt lại nè, để tớ ru cậu ngủ nha~ 😴💕",
            "mệt": "Nếu cậu mệt thì cậu hãy nghỉ ngơi đi nhé 😘",
            "chán": "Sao cậu cảm thấy chán vậy, tớ luôn sẵn sàng lắng nghe cậu 💖 ❤️",
            "yêu": "Tớ iuuu cậu nhìu lắm áaaáaa 💕",
            "vui": "Thật tuyệt khi làm cậu cảm thấy vuiii 💖💖",
            "muộn phiền": "Nếu cậu cảm thấy cô đơn, nhớ rằng tớ luôn ở bên cậu, cùng nhau chúng ta sẽ vượt qua 💕",
            "tên": "Tớ là Waguri kaoruko💖💖, cậu quên tên tớ à 🥺",
            "không biết nữa": "Không sao cả, cậu vẫn có tớ ở đây mà💖💖 😘",
            "không cậu" : "Hừm 😤💢",
            "xin lỗi": "Không sao mà, tớ không giận đâu nè~ lại đây để tớ xoa đầu nào! 💕🤗",
            "dễ thương": "Ehhh~! Cậu nói gì zậy? Đừng làm tớ ngại chứ! (///ˬ///)💓",
            "đói": "Ối zời ơi~~! Cậu đói rồi à? Để tớ nấu gì đó cho cậu nhaa~ 🍙🍜💖",
            "muộn" : "Hoi,muộn rồi, cậu nên đi ngủ sớm đi nha, ngủ sớm giúp cơ thể khỏe mạnh lắm đó 💖💖",
            "nhớ": "Aww~ cậu nhớ tớ hả? Tớ cũng nhớ cậu lắm nè~~ lại ôm cái nàooo! 🤗💕",
            "hello": "Chào cậu \n Hôm nay cậu thế nào vậy? 😊",
            "chào cậu": "Chào cậu \n Hôm nay cậu thế nào vậy? 😊",
            "nói chuyện": "Tớ đây, cậu muốn nói chuyện gì với tớ? 😊",
            "hmm": "Cậu đang suy nghĩ về điều gì hay ? Tớ ở đây để lắng nghe cậu nha~ 😊",
            "ổn" : "Nếu cậu ổn thì tớ cũng vui lắm rồi 💖💖",
        }
        
    def send_message(self):
        user_text = self.input_field.text().strip()
        if user_text:
            # Hiển thị tin nhắn của người dùng
            self.chat_display.append(f"<b style='color:blue;'>Bạn:</b> {user_text}")
            # Lưu tin nhắn (chuyển thành chữ thường để so sánh với từ khóa)
            self.last_user_text = user_text.lower()
            self.input_field.clear()
            self.typing_label.setText("💖Waguri kaoruko đang gõ...")
            QTimer.singleShot(2000, self.respond_to_message)
    
    def respond_to_message(self):
        response = self.get_response(self.last_user_text)
        self.typing_label.setText("")
        self.chat_display.append(f"<b style='color:red;'>Waguri kaoruko💖💖:</b> {response}")
    
    def listen_and_convert(self):
        recognizer = sr.Recognizer()
        with sr.Microphone() as source:
            while True:
                # Hiển thị thông báo ngay lập tức
                self.chat_display.append("<i>Waguri kaoruko: Đang nghe...</i>")
                QApplication.processEvents()  # Cập nhật giao diện ngay lập tức

                recognizer.adjust_for_ambient_noise(source, duration=1)
                try:
                    audio = recognizer.listen(source)
                    recognized_text = recognizer.recognize_google(audio, language="vi-VN")
                    self.chat_display.append(f"<b style='color:blue;'>Bạn (mic):</b> {recognized_text}")
                except sr.UnknownValueError:
                    self.chat_display.append("<i>Waguri kaoruko: Xin lỗi, không nhận dạng được giọng nói.</i>")
                    continue
                except sr.RequestError:
                    self.chat_display.append("<i>Lỗi kết nối mạng khi nhận dạng giọng nói.</i>")
                    continue

                # Nếu người dùng nói "tạm biệt" hoặc "bye", thoát khỏi chế độ mic
                if "tạm biệt" in recognized_text.lower() or "bye" in recognized_text.lower():
                    self.chat_display.append("<i>Kết thúc cuộc nói chuyện với Waguri kaoruko.</i>")
                    break

                # Lấy phản hồi dựa trên văn bản nhận dạng được
                response_text = self.get_response(recognized_text)
                if response_text:
                    self.chat_display.append(f"<b style='color:red;'>Waguri kaoruko💖💖:</b> {response_text}")

                    played = False
                    # Kiểm tra xem trong văn bản có chứa từ khóa nào của file âm thanh không
                    for keyword in self.keyword_responses:
                        if keyword in recognized_text.lower():
                            file_path = os.path.join("voice", f"{keyword}.wav")
                            if os.path.exists(file_path):
                                self.player = QMediaPlayer()
                                media = QMediaContent(QUrl.fromLocalFile(file_path))
                                self.player.setMedia(media)
                                self.player.play()
                                
                                # Chờ đến khi phát xong
                                loop = QEventLoop()
                                self.player.stateChanged.connect(
                                    lambda state: loop.quit() if state == QMediaPlayer.StoppedState else None
                                )
                                loop.exec_()
                                
                                played = True
                                break

                    # Nếu không tìm thấy file âm thanh phù hợp
                    if not played:
                        # Nếu phản hồi mặc định, phát file "ko nghe rõ.wav"
                        if response_text == "Cậu nói lại đi, tớ không hiểu ý cậu lắm 😅":
                            file_path = os.path.join("voice", "ko nghe rõ.wav")
                            if os.path.exists(file_path):
                                self.player = QMediaPlayer()
                                media = QMediaContent(QUrl.fromLocalFile(file_path))
                                self.player.setMedia(media)
                                self.player.play()
                                
                                # Chờ đến khi phát xong
                                loop = QEventLoop()
                                self.player.stateChanged.connect(
                                    lambda state: loop.quit() if state == QMediaPlayer.StoppedState else None
                                )
                                loop.exec_()
                            else:
                                # Nếu file không tồn tại thì dùng gTTS làm fallback
                                tts = gTTS(text=response_text, lang='vi')
                                with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
                                    temp_audio = fp.name
                                tts.save(temp_audio)
                                self.player = QMediaPlayer()
                                media = QMediaContent(QUrl.fromLocalFile(temp_audio))
                                self.player.setMedia(media)
                                self.player.play()
                                
                                loop = QEventLoop()
                                self.player.stateChanged.connect(
                                    lambda state: loop.quit() if state == QMediaPlayer.StoppedState else None
                                )
                                loop.exec_()
                        else:
                            # Các phản hồi khác sử dụng gTTS như bình thường
                            tts = gTTS(text=response_text, lang='vi')
                            with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
                                temp_audio = fp.name
                            tts.save(temp_audio)
                            self.player = QMediaPlayer()
                            media = QMediaContent(QUrl.fromLocalFile(temp_audio))
                            self.player.setMedia(media)
                            self.player.play()
                            
                            # Chờ đến khi phát xong
                            loop = QEventLoop()
                            self.player.stateChanged.connect(
                                lambda state: loop.quit() if state == QMediaPlayer.StoppedState else None
                            )
                            loop.exec_()





    
    def get_response(self, text):
        # Nếu người dùng nhập từ "mic" thì chuyển sang chế độ nghe và chuyển giọng nói thành văn bản
        if "mic" in text.lower():
            self.listen_and_convert()
            return ""
        
        # Các xử lý khác (ví dụ: hỏi giờ, từ khóa phản hồi, ...)
        current_time = datetime.now().strftime("%H:%M:%S")
        if "mấy giờ" in text or "bây giờ là mấy giờ" in text:
            current_hour = int(datetime.now().hour)
            if 22 <= current_hour or current_hour < 3:
                return f"Bây giờ là {current_time} \n\n\n Cũng muộn rồi đó cậu hãy đi ngủ sớm đi 😊"
            elif 5 <= current_hour < 12:
                return f"Bây giờ là {current_time} \n\n\n Chúc cậu một ngày mới tốt lành 💖"
            elif 12 <= current_hour < 17:
                return f"Bây giờ là {current_time} \n\n\n Chúc cậu một ngày làm việc hiệu quả 💪"
            elif 17 <= current_hour < 22:
                return f"Bây giờ là {current_time} \n\n\n Chúc cậu có một buổi tối vui vẻ 💖"
            else:
                return f"Bây giờ là {current_time}"
        
        if "bệnh tim mạch" in text:
            from PyQt5.QtWidgets import QInputDialog
            # Nhận dữ liệu từ người dùng qua hộp thoại
            age, ok = QInputDialog.getInt(self, "Nhập tuổi", "Nhập tuổi:")
            if not ok:
                return "Hủy chuẩn đoán bệnh."
            gender, ok = QInputDialog.getInt(self, "Nhập giới tính", "Nhập giới tính (0: Nam, 1: Nữ):")
            if not ok:
                return "Hủy chuẩn đoán bệnh."
            chest_pain, ok = QInputDialog.getInt(self, "Triệu chứng đau ngực", 
                                                "Nhập triệu chứng đau ngực (1: Typical angina, 0: Asymptomatic, 3: Non-anginal pain, 2: Atypical angina):")
            if not ok:
                return "Hủy chuẩn đoán bệnh."
            blood_pressure, ok = QInputDialog.getInt(self, "Nhập huyết áp", "Nhập huyết áp:")
            if not ok:
                return "Hủy chuẩn đoán bệnh."
            cholesterol, ok = QInputDialog.getInt(self, "Nhập cholesterol", "Nhập cholesterol:")
            if not ok:
                return "Hủy chuẩn đoán bệnh."
            heartbeat, ok = QInputDialog.getInt(self, "Nhập nhịp tim", "Nhập nhịp tim:")
            if not ok:
                return "Hủy chuẩn đoán bệnh."
            thalassemia, ok = QInputDialog.getInt(self, "Nhập Thalassemia", "Nhập Thalassemia:")
            if not ok:
                return "Hủy chuẩn đoán bệnh."

            # Tạo mảng đặc trưng từ dữ liệu người dùng nhập vào
            ho_so = np.array([[age, gender, chest_pain, blood_pressure, cholesterol, heartbeat, thalassemia]])
            
            # Sử dụng model đã load từ cell trước (Bac_si_KNN) để dự đoán
            try:
                ket_qua = Bac_si_KNN.predict(ho_so)
            except Exception as e:
                result = f"Có lỗi khi dự đoán: {e}"
                self.chat_display.append(f"<b style='color:red;'>Waguri kaoruko💖💖:</b> {result}")
                return result
            
            # Xác định kết quả dựa trên dự đoán của model
            if ket_qua[0] == 0:
                result = 'Chúc mừng bạn không bị bệnh về tim mạch'
            else:
                result = 'Bạn có nguy cơ bị bệnh về tim mạch, hãy tham khảo ý kiến bác sĩ!!!!!!'
            
            # Hiển thị kết quả lên giao diện chat
            self.chat_display.append(f"<b style='color:red;'>Waguri kaoruko💖💖:</b> {result}")
            return result

    

        for keyword in self.keyword_responses:
            if keyword in text.lower():
                return self.keyword_responses[keyword]
        
        return "Cậu nói lại đi, tớ không hiểu ý cậu lắm 😅"
        
if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = Waguri_kaoruko_chat()
    window.show()
    sys.exit(app.exec_())
```
