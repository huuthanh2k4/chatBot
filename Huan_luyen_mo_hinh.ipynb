```python
# tỉ lệ giữa người ko bị bệnh và bị bệnh
ti_le_result = data2['Result'].value_counts()
print(ti_le_result)
lb = ['bị bệnh', 'ko Bị bệnh']
# Biểu đồ tròn 
plt.figure(figsize=(10,6))
plt.pie(ti_le_result, labels=lb, autopct='%1.1f%%', wedgeprops=dict(width=0.7, edgecolor='w'))
plt.title('Biểu đồ tròn : Tỉ lệ người bị bệnh và không bị bệnh')
plt.show()

# biểu đồ cột
plt.figure(figsize=(10,6))
plt.bar(lb, ti_le_result, color=['red','skyblue',])
plt.title('Biểu đồ cột : thể hiện số người bị bệnh và ko bị bệnh')
plt.grid(ls='--',lw=1)
plt.show()
```

```python
# Lấy giá trị đọc lập của X
x = data2.loc[:,['Age','Gender','Type','Blood_pressure','Cholesterol','Heartbeat','Thalassemia']]

print('Kích thước biến độc lập X:' ,x.shape)
x
```

```python
# lấy giá trị độc lập của y
y = data2.loc[:,['Result']]

print('Kích thước biến độc lập y:', y.shape)
y
```

```python
# biểu đồ cột và biểu đồ tròn thể hiện số người mắc bệnh và ko mắc bệnh
kiem_tra = y.value_counts()

# vẽ biểu đồ
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.bar(lb,kiem_tra.values,color=['red','skyblue'])
plt.title('Biểu đồ cột thể hiện số người mắc bệnh và ko mắc bệnh')
plt.xlabel('Trạng thái')
plt.ylabel('Số người')


plt.subplot(1,2,2)
plt.pie(kiem_tra.values, labels=lb, colors=['red', 'skyblue'], autopct='%1.1f%%')
plt.title('Biểu đồ tròn thể hiện số người mắc bệnh và ko mắc bệnh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

huan_luyen_x,text_x,huan_luyen_y,test_y = train_test_split(x,
                                                           y,
                                                           train_size=0.8,
                                                           shuffle=True)
print('Kích thước tập huấn luyện:', huan_luyen_x.shape, huan_luyen_y.shape)
print('Kích thước tập kiểm tra: ', text_x.shape, test_y.shape)
```

```python
print('Kích thước biến độc lập của tập huấn luyện x:', huan_luyen_x.shape)
huan_luyen_x
```

```python
print('Kích thước biến phụ thuộc y: ',huan_luyen_y.shape)
huan_luyen_y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
# chuẩn hóa dữ liệu cho mô hình

Chuan_hoa_data = StandardScaler()
Chuan_hoa_TapHL_x = Chuan_hoa_data.fit_transform(huan_luyen_x)
Chuan_hoa_Test_x = Chuan_hoa_data.transform(text_x)

# Đặt tên cho mô hình

KNN = KNeighborsClassifier(n_neighbors=11,
                           weights='distance',
                           algorithm='ball_tree',
                           p=1)

# Huấn luyện mô hình đã chuẩn hóa
KNN.fit(Chuan_hoa_TapHL_x,huan_luyen_y)
print(KNN)

Do_chinh_xac = KNN.score(Chuan_hoa_TapHL_x,huan_luyen_y)
print('Độ chích xác của mô hình là: ', round(Do_chinh_xac*100),'%')
```

```python
from sklearn.model_selection import GridSearchCV

Tham_so = {
    'n_neighbors': range(1, 21),
    'weights': ['uniform', 'distance'],
    'p': [1, 2],
    'algorithm': ['ball_tree', 'kd_tree', 'auto']
}

Search_luoi = GridSearchCV(KNeighborsClassifier(),Tham_so,cv=5,scoring='accuracy')
Search_luoi.fit(Chuan_hoa_TapHL_x,huan_luyen_y)

print('Tham số tốt nhất: ', Search_luoi.best_params_)
print('Độ chính xác khi qua kiểm định: ', Search_luoi.best_score_*100,'%')
```

```python
from sklearn.metrics import accuracy_score

# Đánh giá độ chính xác của mô hình

Kq_du_doan = KNN.predict(Chuan_hoa_Test_x)

Do_chinh_xac_kq = accuracy_score(test_y,Kq_du_doan)
print('Độ chính xác của mô hình trên tập kiểm tra là: ', Do_chinh_xac_kq*100, '%')
```

```python
kq_true = accuracy_score(test_y, Kq_du_doan, normalize=False)
print('Số lượng dự đoán đúng trên tập kiểm tra: ', kq_true, '/', len(text_x))
```
